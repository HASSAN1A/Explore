### Extracting Insights from Mentions

In this section we will seek to display the top 10 mentions of each sentiment by count and by chi2-score in tabular and visual format. To do this we will first have to extract all the mentions into a separate column, create and use a bag of words function per sentiment and work out the percentage each mention is used over the total mentions in the dataset.

Extracting all the unique sentiments and placing them in a seperate column.

```python
df_train
```

```python
unique_sentiments = list(df_train['sentiment'].unique())
unique_sentiments
```

Creating a bag of words function  which counts the amount of time each word occurs. This function will be used to count the amount of mentions, hashtags and tokens per sentiment.

```python
def bag_of_words_count(words, word_dict={}):
    """ this function takes in a list of words and returns a dictionary 
        with each word as a key, and the value represents the number of 
        times that word appeared"""
    for word in list(words.split()):
        if word in word_dict.keys():
            word_dict[word] += 1
        else:
            word_dict[word] = 1
    return word_dict
```

Extracting all the mentions from the message column so we can work with them.

```python
# Extracting all mentions into a seperate column and making nan values strings for smooth operation.
df_train['mentions'] = df_train['message'].str.extractall(
    '(\@\w+.*?)').groupby(level=0)[0].apply(' '.join)
df_train['mentions'] = df_train['mentions'].replace(np.nan, '')
df_train['mentions']
```

Counting the times each mention is used by sentiment using the bag of words function.

```python
# Counting the amount each mention occurs for each sentiment
sentiment = {}
for pp in unique_sentiments:
    df = df_train.groupby('sentiment')
    sentiment[pp] = {}
    for row in df.get_group(pp)['mentions']:
        sentiment[pp] = bag_of_words_count(row, sentiment[pp])
sentiment
```

Creating a combined bag of words dictionary for all the mentions in the dataset so that we can see the total count per word.

```python
# Creating a set of all unique words.
all_mentions = set()
for pp in unique_sentiments:
    for word in sentiment[pp]:
        all_mentions.add(word)
```

```python
# Creating a combined bag of words dictionary for all the mentions in the text.
sentiment['Total mentions'] = {}
for pp in unique_sentiments:
    for word in all_mentions:
        if word in sentiment[pp].keys():
            if word in sentiment['Total mentions']:
                sentiment['Total mentions'][word] += sentiment[pp][word]
            else:
                sentiment['Total mentions'][word] = sentiment[pp][word]
```

```python
# Calculating the total amount of mentions.
total_mentions = sum([v for v in sentiment['Total mentions'].values()])
total_mentions
```

Here we will work out the mention counts by sentiment as well as the total mention counts and display only the top 10 of the total mentions. We will also change the values 1, 2, 0 and -1 to a better description of the sentiment.

```python
# Creating a dataframe with the word counts by sentiment.
hm = []
for p, p_bow in sentiment.items():
    df_bow = pd.DataFrame([(k, v) for k, v in p_bow.items(
    ) if k in sentiment['Total mentions']], columns=['Word', p])
    df_bow.set_index('Word', inplace=True)
    hm.append(df_bow)

# create one big dataframe
df_bow = pd.concat(hm, axis=1)
df_bow.fillna(0, inplace=True)
```

```python
# Changing the names of the columns to be clearer descriptions of the sentiment
df_bow = df_bow.rename(columns={1: 'Pro', 2: 'News', 0: 'Neutral', -1: "Anti"})
```

```python
# Finding the top 10 occuring words
df_bow.sort_values(by='Total mentions', ascending=False).head(10)
```

Working out the mentions percentage by sentiment to work out the chi2 score with.

```python
# Calculating the percentage score for each sentiment.
df_bow['News_perc'] = df_bow['News'] / df_bow['News'].sum()
df_bow['Pro_perc'] = df_bow['Pro'] / df_bow['Pro'].sum()
df_bow['Neutral_perc'] = df_bow['Neutral'] / df_bow['Neutral'].sum()
df_bow['Anti_perc'] = df_bow['Anti'] / df_bow['Anti'].sum()
df_bow['Total mentions_perc'] = df_bow['Total mentions'] / \
    df_bow['Total mentions'].sum()
```

Working out the chi2 scores for each sentiment. The chi2 score will show the amount that this specific sentiment differs from the use of the other sentiments.

```python
# Calculating the chi2 score for each sentiment.
df_bow['News_chi2-score'] = np.power((df_bow['News_perc'] -
                                      df_bow['Total mentions_perc']), 2) / df_bow['Total mentions_perc']
df_bow['Pro_chi2-score'] = np.power((df_bow['Pro_perc'] -
                                     df_bow['Total mentions_perc']), 2) / df_bow['Total mentions_perc']
df_bow['Neutral_chi2-score'] = np.power((df_bow['Neutral_perc'] -
                                         df_bow['Total mentions_perc']), 2) / df_bow['Total mentions_perc']
df_bow['Anti_chi2-score'] = np.power((df_bow['Anti_perc'] -
                                      df_bow['Total mentions_perc']), 2) / df_bow['Total mentions_perc']
```

```python
df_bow[['News', 'News_chi2-score']][df_bow['News_perc'] > df_bow['Total mentions_perc']
                                    ].sort_values(by='News_chi2-score', ascending=False).head(10)
```

```python
# Calculating the percentage score for each sentiment.
df_hash['News' + '_perc'] = df_hash['News'] / df_hash['News'].sum()
df_hash['Pro' + '_perc'] = df_hash['Pro'] / df_hash['Pro'].sum()
df_hash['Neutral' + '_perc'] = df_hash['Neutral'] / df_hash['Neutral'].sum()
df_hash['Anti' + '_perc'] = df_hash['Anti'] / df_hash['Anti'].sum()
df_hash['Total hashtags' + '_perc'] = df_hash['Total hashtags'] / \
    df_hash['Total hashtags'].sum()
```

Working out add displaying the top chi2-scores for each sentiment.

```python
# Calculating the chi2 score for each sentiment.
df_hash['News_chi2-score'] = np.power((df_hash['News_perc'] -
                                       df_hash['Total hashtags_perc']), 2) / df_hash['Total hashtags_perc']
df_hash['Pro_chi2-score'] = np.power((df_hash['Pro_perc'] -
                                      df_hash['Total hashtags_perc']), 2) / df_hash['Total hashtags_perc']
df_hash['Neutral_chi2-score'] = np.power((df_hash['Neutral_perc'] -
                                          df_hash['Total hashtags_perc']), 2) / df_hash['Total hashtags_perc']
df_hash['Anti_chi2-score'] = np.power((df_hash['Anti_perc'] -
                                       df_hash['Total hashtags_perc']), 2) / df_hash['Total hashtags_perc']
```

## Text Pre-processing üî†
Text is the least structured of all the data types available; it contains many sorts of noise and is difficult to analyze without pre-processing. Text preprocessing refers to the entire process of cleaning and standardizing text to make it noise-free and suitable for analysis.

This process is predominantly comprised of three steps::

- Tokenization
- Noise Removal
- Lexicon Normalization

### Tokenization
Tokenizing is like splitting a whole sentence into words. You can consider a simple separator for this purpose. But a separator will fail to split the abbreviations separated by "." or special characters, like U.A.R.T., for example. Challenges increase when more languages are included. How about dealing with compound words in languages such as German or French?

Most of these problems can be solved by using the `nltk` library. The `word_tokenize` module breaks the words into tokens and these words act as an input for the normalization and cleaning process. It can further be used to convert a string (text) into numeric data so that machine learning models can digest it.

There is a common approach to lowercasing everything for the sake of simplicity. It helps to maintain the consistency flow during the NLP tasks and text mining. The lower() function makes the whole process quite straightforward. 

We complete both tasks below.

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text


df_train['token'] = df_train['message'].apply(
    lambda x: tokenization(x.lower()))
df_test['token'] = df_test['message'].apply(lambda x: tokenization(x.lower()))
```

Removing **noise** comes in handy when you want to do text analysis on pieces of data like comments or tweets. The code in the following sections will be helpful to get rid of the text that interferes with text analysis

####  Stemming
There are many variations of words that do not bring any new information and create redundancy, ultimately bringing ambiguity when training machine learning models for predictions. Take "He likes to walk" and "He likes walking," for example. Both have the same meaning, so the stemming function will remove the suffix and convert "walking" to "walk."

#### Lemmatization
Unlike stemming, *lemmatization* performs normalization using vocabulary and morphological analysis of words. Lemmatization aims to remove inflectional endings only and to return the base or dictionary form of a word, which is known as the *lemma*. Lemmatization uses a dictionary, which makes it slower than stemming, however the results make much more sense than what you get from stemming. Lemmatization is built on WordNet's built-in morphology function, making it an intelligent operation for text analysis. A WordNet module is a large and public lexical database for the English language. Its aim is to maintain the structured relationship between the words. The `WordNetLemmitizer()` is the earliest and most widely used function.

Although, we will just use lematization:

```python
# For train dataset
wn = nltk.WordNetLemmatizer()


def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text


df_train['lem'] = df_train['stop'].apply(lambda x: lemmatizer(x))


wn = nltk.WordNetLemmatizer()

# For test dataset


def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text


df_test['lem'] = df_test['stop'].apply(lambda x: lemmatizer(x))

df_train.head()
```

### Wordclouds
Word clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

Below we create a function which can work out the appropriate metadata included and display it in word cloud format.

```python
train = pd.read_csv('https://github.com/RickieMohale/classification-predict-streamlit-template/blob/master/train.csv?raw=true')
# Converting tweets to lowercase
train['message'] = train['message'].str.lower()
string.punctuation


def clean(text):
    text = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)

    words = text.split()

    return(" ".join(words))


train['message'] = train['message'].apply(clean)


# Remove Stop words
def stop(text):
    word = text.split()
    # Remove stop words
    stop_4twit = ['rt', 'rts', 'retweet']
    stop_word = set().union(stopwords.words('english'), stop_4twit)
    remove_stop = [w for w in word if w not in stop_word]
    nexstop = " ".join(remove_stop)

    return nexstop


train['message'] = train['message'].apply(stop)


def lemmatizer(text):

    lemma = WordNetLemmatizer()
    new_text = " ".join([lemma.lemmatize(lem) for lem in text.split()])

    return new_text


train['message'] = train['message'].apply(lemmatizer)
```

### Hyperparameter Tuning
Here we use a pipeline to tune our model to run both TF-IDf and Count Vectorizer at the same time, with specific parameters. We also use Grid Search which builds a model for every combination of hyperparameters specified and evaluates each model

```python
# Creating a pipeline for the gridsearch
param_grid = {'C': [0.01, 0.03, 0.06, 0.09]}  # setting parameter grid

tuned = Pipeline([('tfidf', TfidfVectorizer(min_df=1,
                                            max_df=0.9,
                                            ngram_range=(1, 2))),
                  ('mnb', GridSearchCV(LinearSVC(),
                                       param_grid=param_grid,
                                       cv=10,
                                       n_jobs=-1,
                                       scoring='f1_weighted'))
                  ])
tuned.fit(X_train, y_train)
y_pred = tuned.predict(X_test)
f1_macro = metrics.f1_score(y_test, y_pred, average='macro')
accuracy = metrics.f1_score(y_test, y_pred, average='micro')

experiment.log_dataset_hash(X_train)
experiment.log_parameters({"model_type": "Tuned", "param_grid": param_grid})
experiment.log_metrics({'F1 Macro': f1_macro, "Accuracy": accuracy})
experiment.end()
```
